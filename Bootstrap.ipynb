```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 1000
means = np.zeros(N)
for i in range(N):
  draws = np.random.wald(15,1000, size= 10)
  mean = np.mean(draws)
  means[i] = mean
plt.hist(means)
```

En realidad yo medi esto

```python
draws = np.random.wald(15, 1000, size=200)
plt.hist(draws)
```

Bootstrap

```python
N = 10000
bootstrap_means = np.zeros(N)
for i in range(N):
  mean = np.mean(np.random.choice(draws, size=len(draws), replace= True)) # size debe ser el mismo que en el muestreo original
  bootstrap_means[i] = mean
plt.hist(bootstrap_means)
```

No usar el bootstrap si tenemos dudas sobre la distribucion empirica.  
Si el bootstrap falla es dificil ver donde.


```python
sem = np.std(bootstrap_means)
sem # error estandar de la media
```

```python
np.std(draws)/np.sqrt((len(draws) - 1))
```

Intervalos de confianza:

```python
print(np.quantile(bootstrap_means, 0.025),
np.quantile(bootstrap_means, 0.975))
```

bootstrap para std

```python
N = 10000
bootstrap_stds = np.zeros(N)
for i in range(N):
  std = np.std(np.random.choice(draws, size=len(draws), replace= True)) # size debe ser el mismo que en el muestreo original
  bootstrap_stds[i] = std
plt.hist(bootstrap_stds)
```
