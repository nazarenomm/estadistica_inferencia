Tirar una moneda $N$ veces y salieron $y$ caras. Queremos averiguar cosas relativas a la probabilidad de que salga cara.  

Para resolver esto usabamos EMV (MLE)  
Como es un problema de una moneda uso bernoulli



Encontremos el valor de $p$ que maximiza $L$



La likelihood es casi una probabilidad pero no integra a 1

Para solucionarlo usamos el _Teorema de Bayes_

$P(p|y,n) = \frac{P(y,n|p)P(p)}{P(y,n)}$

$p$ es un parametro (podra haber sido un $\theta$), es raro que aparezca la probabilidad de un parametro.



A la probabilidad total de la data $P(y,n)$, no vale la pena analizarla.

Prior: distribucin de probabilidad sobre el/los parmetros

El prior se elige basandose en la inforamcion conocida sobre el problema con anterioridad.  
Por ejemplo: estoy seguro de que $p > 0.5$



Otro ejemplo: estoy seguro de $p = 0.5$

 (distribucin puntual en 0.5)

```python
from scipy.stats import beta
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 1001)
y = beta.pdf(x, a= 31, b=71)
y2 = beta.pdf(x, a=4, b= 8) # 3 casos favorables(a -1) sobre 10 (a + b -2)
fig, ax = plt.subplots()
ax.plot(x,y, label = "30/100")
ax.plot(x, y2, label = "3/10")
ax.legend()
```
