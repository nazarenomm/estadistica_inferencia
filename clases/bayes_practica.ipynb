La aproximacion normal se usa cuando quiero una aproximacion simple que tenga en cuenta TODA la informacion.
Se usa cuando la posterior no es muy sesgada, no muy asimetrica, no tenga colas pesadas (que no se hagan mas chicas cuadno nos alejamos del centro)

Vamos a hacer el ejercicios de la pr√°ctica 7



$y:$ cantidad de $6$

$y \sim Bin(p=\frac{1}{6}, n =1000)$
Para aproximar $P(y)$ necesito:
+ $E[y] = np = \frac{1000}{6}$
+ $Var[y] = np(1-p) = 1000 .\frac{1}{6}.\frac{5}{6}$
  
$\Rightarrow y \approx N(\frac{1000}{6}, \frac{1000}{6}.\frac{5}{6})$
