```python
x1 = np.random.normal(0,1,100)
x2 = np.random.normal(0,1,100)
x3 = np.random.normal(0,1,100)

z = np.random.normal(0,0.25**.5,100)

y = 3*x1 + 2*x2 + x3 + z
```

```python
import numpy as np

# Assuming X and Y are numpy arrays
# X is a 2D array where each row is a sample and each column is a feature
# Y is a 1D array of the target variable for each sample
unos = np.ones(100)

X = np.array([x1,x2,x3,unos]).T

# Add a column of ones to X for the intercept term

# Compute the least squares estimator
beta = np.linalg.inv(X.T @ X) @ X.T @ y
```

```python
rep_exp = 3000
betas = []

var = []

for i in range(rep_exp):
  x1 = np.random.normal(0,1,100)
  x2 = np.random.normal(0,1,100)
  x3 = np.random.normal(0,1,100)

  z = np.random.normal(0,1,100)

  y = 3*x1 + 2*x2 + x3 + z
  X = np.array([x1,x2,x3,unos]).T

  beta = np.linalg.inv(X.T @ X) @ X.T @ y

  betas.append(beta)
  y_pred = X@beta




  sigma_2 = np.array((y-y_pred)**2).sum()/(100-3-1)
  var.append(sigma_2 * np.linalg.inv(X.T @ X))

betas
```

```python
b1 = [ x[0] for x in betas]
plt.hist(b1, bins = 20)
plt.title("B1")
```
