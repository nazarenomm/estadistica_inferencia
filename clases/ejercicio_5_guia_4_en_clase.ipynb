```python
from scipy.stats import binom
import matplotlib.pyplot as plt
n, p= 1000, 0.5
binomial = binom(n, p)
```

```python
c = 501
```

```python
def tail_binomial(c):
    return sum([binomial.pmf(i) for i in range(c, 1001)])
```

```python
tails = []
values = range(1001)

for c in values:
    tails.append(tail_binomial(c))
```

```python
fig, ax = plt.subplots()
ax.plot(values, tails)
ax.grid()
ax.axhline(y = 0.5, c = "r")
plt.show()
```
